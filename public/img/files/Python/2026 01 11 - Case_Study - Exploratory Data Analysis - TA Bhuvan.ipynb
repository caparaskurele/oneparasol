```python
for col in df.columns:
  if(df[col].dtypes == "object"):
    mode1 = df[col].mode()[0]
    df[col] = df[col].fillna(mode1)
  else:
    mean1 = df[col].mean()
    df[col] = df[col].fillna(mean1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in df.columns:
  if(df[col].dtypes != "object"):
    sns.boxplot(df[col])
    plt.title(col)
    plt.show()
```

```python
# Remove the Outliers
# IQR

# time -> 12:36
```

```python
# Q1
# Q3
# IQR = Q3-Q1
# LB = Q1-1.5*IQR
# UB = Q3+1.5*IQR
# df[col] >= LB & df[col] <= UB
```

```python
for col in df.columns:
  if(df[col].dtypes != "object"):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1

    LB = Q1-1.5*IQR
    UB = Q3+1.5*IQR

    df = df[(df[col]>=LB) & (df[col]<=UB)]
```

```python
for col in df.columns:
  if(df[col].dtypes != "object"):
    sns.boxplot(df[col])
    plt.title(col)
    plt.show()
```
