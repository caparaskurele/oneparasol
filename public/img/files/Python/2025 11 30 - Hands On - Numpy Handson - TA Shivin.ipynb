9) Create a array of numbers, and filter out all the numbers which are greater than the avg of the array.


 Don't use for loop for filter.



```python
arr = np.array([10, 25, 30, 5, 40, 15, 35, 20, 45, 8])
avg=arr.mean()
print("Avg=",avg)

ans=arr[arr>avg]
print(ans)
```

10. Creat an array of nuumber, filter out all the even numbers.

```python
arr = np.array([10, 25, 30, 5, 40, 15, 35, 20, 45, 8])

ans=arr[arr % 2 ==0]

print(ans)
```
