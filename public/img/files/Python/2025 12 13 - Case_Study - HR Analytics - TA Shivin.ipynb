# Visualizaton

**1) Make the distrbution plot of the statistaction levels, have 10 bins.**

Give xlabel, ylabel, and title

```python
plt.hist(df['satisfaction_level'],bins=10,rwidth=0.95)
plt.xlabel("Satisfaction")
plt.ylabel("Number of employees")
plt.title("Distribution of Satisfaction level")
plt.show()
```

**2) Create a plot which tells us the count of the number of employess with each salary level( low, mid, high).**



```python
sb.countplot(data=df,x="salary",color="lightblue")
plt.show()
```

**3) Create a pie chart, for left column, tell the protion of left the company or not left.**

```python
count=df["left"].value_counts()

print(count)

plt.pie(count,labels=["Not left","Left"],autopct="%1.2f%%")
plt.show()
```

**4) Find the relationship between, satisfaction level vs monthly hours**

```python
sb.scatterplot(data=df,x='satisfaction_level',y='average_montly_hours')
plt.show()
```

**7) Using a plot tell how many people have got promotion in last 5 years, and how many have not.**

```python
plt.figure(figsize=(4, 3))
plt_7=sb.countplot(data=df,x="promotion_last_5years")
# plt.show()

for i in plt_7.containers:
  plt_7.bar_label(i)

plt.show()
```
