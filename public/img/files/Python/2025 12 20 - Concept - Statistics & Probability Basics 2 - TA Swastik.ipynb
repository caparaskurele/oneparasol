Variance  : It quantifies the spread of data around mean

image.png

```python
import numpy as np
import statistics as stats
```

```python
data = [25,30,35,40,45,50,55,60,65,70]
print(sum(data))
ans = np.var(data)
print(ans)
```

```python
ans = stats.variance(data)
print(ans)
```

```python
print(sum(data)/len(data))
```

```python
dev = 0
mean = sum(data)/len(data)
for i in data:
  dev += (i-mean)**2
print(dev)
```

```python
sam_var = dev / (len(data)-1)
```

```python
print(sam_var)
```

Standard Deviation : The square root of variance

image.png

```python
#10,12,14,16,18
```

```python
data = [10,12,14,16,18]
pop_var = np.var(data)
print(pop_var)
```

Persentile -> A percentile indicates the percentage of data values that fall at or below the percentile point

```python
#Percetage -> 95% => I got 95 marks out of 100
#Percentile => 95 => 95% percetage of students are equal or below than me
```

```python
data = [10,30,64,24,63,73,78,35,57,46]

#60 percentile
ans = np.percentile(data,60)
print(ans)
```

```python
#Quartiles ->
# 1. First Quartile(Q1) : 25 percentile
# 2. Second Quartile(Q2) : 50 percentile
# 3. Third Quatile(Q3) : 75 percentile
```

```python
data = [10,30,64,24,63,73,78,35,57,46]
q1 = np.percentile(data,25)
q2 = np.percentile(data,50)
q3 = np.percentile(data,75)
print(q1,q2,q3)
```

```python
#Outliers => An outliers is a data point that is very differnet from the rest of the datapoints
 #Extremely High, Extremely low
```

```python
import matplotlib.pyplot as plt
```

```python
#outlier detection
age = [20,24,35,250,46,58,57,38,49,270,-230,-180]
plt.boxplot(age)
plt.show()
```

```python
import seaborn as sns
sns.boxplot(age)
plt.show()
```

```python
#Outlier Removal
```

```python
# Inter Quartile range
#IQR = Q3 - Q1
```

```python
Q1 = np.percentile(data,25)
Q3 = np.percentile(data,75)
```

```python
IQR = Q3 - Q1
```

```python
UL = Q3 + 1.5*IQR
LL = Q1 - 1.5*IQR
```

```python
new_age = []
for i in age:
  if i<UL and i>LL:
    new_age.append(i)
```

```python
print(new_age)
```

```python
plt.boxplot(new_age)
plt.show()
```
