```python
import seaborn as sb

data=sb.load_dataset('penguins')
print(data.head())
```

**2) Plot a histogram of the penguin flipper lengths, bins=15**

```python
plt.figure(figsize=(7,3))
sb.histplot(data=data,x="flipper_length_mm",bins=15,color='teal',edgecolor='darkblue')
plt.title("Distribution of flipper length")
plt.show()
```

**3) Pie char for proportion of penguin species( number of penguins in each species)**

**Use matplolib**

```python
count=data["species"].value_counts()
plt.figure(figsize=(3,3))
plt.pie(count,labels=count.index,autopct="%0.2f%%")
plt.show()
```

**4) Show the relationship between Flipper length and body mass.**

```python
sb.lineplot(data=data,x="flipper_length_mm",y="body_mass_g",hue="species")
plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length")
plt.ylabel("Body Mass")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sb.scatterplot(data=data,x='body_mass_g',y='flipper_length_mm',hue="species")
plt.show()
```

**5) Find the avg body mass of each species( use groupby), And then create a bar plot.**

```python
avg=data.groupby("species")["body_mass_g"].mean()

sb.barplot(x=avg.index,y=avg,width=0.3,palette=["orange","lightblue","limegreen"])
plt.show()
```

```python
print(list(data.columns))
```

**Q6) Find the avg bill_length_mm of each island and show using horizontal bar plot**

```python
avg=data.groupby("island")['bill_length_mm'].mean()
sb.barplot(x=avg,y=avg.index,width=0.3)
plt.show()
```

**7) Check for the dstribution and outlier in body mass using box plot , for each Species**

```python
plt.figure(figsize=(5,3))
sb.boxplot(data=data,y='body_mass_g',hue="species")
plt.show()
```
