```python
A
```

```python
# Time -> 11:20
```

```python
A = np.array([[4,5,1],[3,8,4],[20,3,6]])
B = np.array([[12,14,5],[45,5,5],[9,8,3]])
```

```python
A
```

```python
B
```

```python
# L.H.S
lhs = (A+B).T
print(lhs)

# R.H.S
rhs = A.T + B.T
print(rhs)

if np.array_equal(lhs, rhs):
  print("Additive property is Correct")
else:
  print("Wrong")
```

```python
# Property 3 -> Multiplicative Property
# (AxB).T = B.T x A.T

# time -> 11:34
```

```python
# L.H.S
lhs = np.dot(A,B).T
print(lhs)

# R.H.S
rhs = B.T.dot(A.T)
print(rhs)

if (np.array_equal(lhs, rhs)):
  print("Multiplicative Property is True")
else:
  print("Multiplicative Property is False")
```

```python
# Check if a matrix is invertible or a singular matrix
# Singular -> Determinant == 0
# Invertible -> Deteminant != 0

# time -> 12:20
```

```python
A = np.random.randint(1,10, size=(3,3))
A
```

```python
detA = np.linalg.det(A)
if detA==0:
  print("A is Singular")
else:
  print("A is Invertible")
```

```python
# Find the inverse of a Matrix
# A * A-1 = Indentity Matrix
# Prove it also

# time -> 12:34
```

```python
M = np.array([[1,2],[3,4]])
M
```

```python
M_inv = np.linalg.inv(M)
M_inv
```

```python
result = np.dot(M, M_inv)
```

```python
result
```

```python
# np.allclose() -> Is a function in numpy to check if two matrix are equal or not...
# if there are some floating value or tiny errors then also it will consider both as same
if np.allclose(result, np.identity(2)):
  print("True")
else:
  print("False")
```
