```python
# Steps ->
# Step1 -> Data Collection
# Step2 -> Prepare Sample
# Step3 -> Creating two Hypothesis (Null, and Alternate)
# Step4 -> Apply the test
# Step5 -> Check p_value ->
          # p_value > 0.05 -> Failed to Reject the Null Hypothesis
          # p_value < 0.05 -> Reject the Null hypothesis
```

```python
# It was observed that the average purchase made by the men of the age 18 to 25 was 10000
# Is it still the Same?
```

```python
new_df = df[(df["Gender"] == 1) & (df["Age"]==1)]
```

```python
df.shape
```

```python
new_df
```

```python
# Sample ->
sample = new_df.sample(3600, random_state=0)
sample
```

```python
# Null Hypothesis -> Mean will be 10000
# Alternate Hypothesis -> Mean will not be equal to 10000
```

```python
# One Sample T-Test
from scipy.stats import ttest_1samp
```

```python
a_mean = 10000
```

```python
t_stat,p_value = ttest_1samp(sample['Purchase'], a_mean)
```

```python
print(t_stat, p_value)
```

```python
if p_value>0.05:
  print("Failed to Reject the Null (Mean is 10000)")
else:
  print("Reject the Null (Mean is not 10000)")
```

```python
# Is the average purchase made by men and women of the age group 18-25 are same?

# time -> 12:50
```

```python
# Step 1 ->
# men, age -> 18-25
# women, age -> 18-25

data_men = df[(df["Age"]==1) & (df["Gender"]==1)]
data_women = df[(df["Age"]==1) & (df["Gender"]==0)]
```

```python
data_men
```

```python
data_women
```

```python
# Step 2 -> Prepare the sample
data_men_sample = data_men.sample(3600, random_state=0)
data_women_sample = data_women.sample(1100, random_state=0)
```

```python
# Step 3  ->
# Null -> average purchase made by men and women of the age group 18-25 are same
# Alternate -> average purchase made by men and women of the age group 18-25 are not same
```

```python
# In two sampled ttest we need to assume that the variance of the two samples are same
```

```python
data_men_sample['Purchase'].var()
```

```python
data_women_sample['Purchase'].var()
```

```python
# Since the variance are diffrent we need to assume it as same
```

```python
from scipy.stats import ttest_ind
```

```python
t_stat, p_value = ttest_ind(data_men_sample['Purchase'], data_women_sample['Purchase'], equal_var=True)
```

```python
print(t_stat, p_value)
```

```python
if p_value>0.05:
  print("Falied to Rejct the Null / Average value of two samples are same")
else:
  print("Reject the Null / Average value of two samples are not same")
```
