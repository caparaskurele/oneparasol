**18) Replace the null values in the numerical columns ( mass, bill length , depth, flipper lenght) with the mean of those columns**

```python
# m=df["mass"].mean()
# df["col"].fillna(m,inplace=True)

num_cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]

df[num_cols]=df[num_cols].fillna(df[num_cols].mean())
```

**19) Replace the na values in categorical cols with mode**

```python
# cols=["species","island","sex"]
df['sex']=df['sex'].fillna(df['sex'].mode()[0])

df['species']=df['species'].fillna(df['species'].mode()[0])

df['island']=df['island'].fillna(df['island'].mode()[0])

num_cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]

df[num_cols]=df[num_cols].fillna(df[num_cols].mean())
print(df)
```
